1. Write a Python program to perform a Z-test for comparing a sample mean to a known population mean and
interpret the results.

```python
import numpy as np
from scipy import stats

def z_test(sample_data, population_mean, population_std, sample_size, significance_level=0.05):
    sample_mean = np.mean(sample_data)
    z_score = (sample_mean - population_mean) / (population_std / np.sqrt(sample_size))
    p_value = 2 * (1 - stats.norm.cdf(abs(z_score)))

    print(f"Sample Mean: {sample_mean}")
    print(f"Z-score: {z_score}")
    print(f"P-value: {p_value}")

    if p_value < significance_level:
        print("Reject the null hypothesis: There is a significant difference between the sample mean and population mean.")
    else:
        print("Fail to reject the null hypothesis: There is no significant difference between the sample mean and population mean.")
```

```python
if __name__ == "__main__":
    sample_data = [85, 88, 91, 83, 79, 94, 88, 82, 91, 87,
                   84, 79, 90, 93, 86, 89, 84, 80, 95, 87,
                   82, 88, 90, 91, 85, 83, 92, 87, 86, 90]

    population_mean = 85
    population_std = 5
    sample_size = len(sample_data)
    z_test(sample_data, population_mean, population_std, sample_size)
```

4. Perform a two-tailed Z-test using Python and visualize the decision region on a plot.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

def two_tailed_z_test(sample_data, population_mean, population_std, significance_level=0.05):
    sample_mean = np.mean(sample_data)
    sample_size = len(sample_data)
    z_score = (sample_mean - population_mean) / (population_std / np.sqrt(sample_size))
    p_value = 2 * (1 - stats.norm.cdf(abs(z_score)))

    print(f"Sample Mean: {sample_mean:.2f}")
    print(f"Z-score: {z_score:.2f}")
    print(f"P-value: {p_value:.4f}")

    if p_value < significance_level:
        print("Reject the null hypothesis: The sample mean is significantly different from the population mean.")
    else:
        print("Fail to reject the null hypothesis: The sample mean is not significantly different from the population mean.")

    plot_decision_region(z_score, significance_level)

def plot_decision_region(z_score, significance_level):
    critical_value = stats.norm.ppf(1 - significance_level / 2)
    x = np.linspace(-4, 4, 1000)
    y = stats.norm.pdf(x)

    plt.plot(x, y, label='Standard Normal Distribution')
    plt.fill_between(x, y, where=(x <= -critical_value) | (x >= critical_value), color='red', alpha=0.3, label='Rejection Region')
    plt.axvline(z_score, color='blue', linestyle='dashed', label=f'Z-score: {z_score:.2f}')

    plt.axvline(critical_value, color='green', linestyle='dashed', label=f'Critical Z-value: ¬±{critical_value:.2f}')
    plt.axvline(-critical_value, color='green', linestyle='dashed')
    plt.title('Two-Tailed Z-Test with Decision Region')
    plt.xlabel('Z-Score')
    plt.ylabel('Probability Density')
    plt.legend(loc='best')
    plt.show()
```

```python
if __name__ == "__main__":
    sample_data = np.random.normal(loc=102, scale=15, size=50)
    population_mean = 100
    population_std = 15

    two_tailed_z_test(sample_data, population_mean, population_std)
```

5. Create a Python function that calculates and visualizes Type 1 and Type 2 errors during hypothesis testing.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

def visualize_type_errors(population_mean, population_std, sample_size, alternative_mean, significance_level=0.05):
    critical_value = stats.norm.ppf(1 - significance_level / 2)
    null_distribution = np.random.normal(loc=population_mean, scale=population_std, size=10000)
    alternative_distribution = np.random.normal(loc=alternative_mean, scale=population_std, size=10000)
    sample_means_null = np.random.normal(loc=population_mean, scale=population_std/np.sqrt(sample_size), size=10000)
    sample_means_alt = np.random.normal(loc=alternative_mean, scale=population_std/np.sqrt(sample_size), size=10000)

    plt.figure(figsize=(10, 6))
    plt.hist(sample_means_null, bins=30, alpha=0.5, label='Null Hypothesis (H‚ÇÄ)', color='blue', density=True)
    plt.hist(sample_means_alt, bins=30, alpha=0.5, label='Alternative Hypothesis (H‚ÇÅ)', color='red', density=True)
    plt.axvline(x=-critical_value, color='green', linestyle='dashed', label=f'Critical Z-value: ¬±{critical_value:.2f}')
    plt.axvline(x=critical_value, color='green', linestyle='dashed')
    plt.fill_betweenx(np.linspace(0, 0.03, 100), critical_value, 4, color='green', alpha=0.3, label='Type I Error Region')
    plt.fill_betweenx(np.linspace(0, 0.03, 100), -4, -critical_value, color='green', alpha=0.3)
    plt.fill_betweenx(np.linspace(0, 0.03, 100), -critical_value, critical_value, color='red', alpha=0.3, label='Type II Error Region')
    plt.title('Visualization of Type I and Type II Errors in Hypothesis Testing')
    plt.xlabel('Sample Mean')
    plt.ylabel('Density')
    plt.legend(loc='best')
    plt.show()
```

```python
if __name__ == "__main__":
    population_mean = 100
    population_std = 15
    sample_size = 30
    alternative_mean = 105

    visualize_type_errors(population_mean, population_std, sample_size, alternative_mean)
```

6.  Write a Python program to perform an independent T-test and interpret the results.

```python
import numpy as np
from scipy import stats

def independent_t_test(sample1, sample2, significance_level=0.05):
    t_statistic, p_value = stats.ttest_ind(sample1, sample2)

    print(f"T-statistic: {t_statistic:.4f}")
    print(f"P-value: {p_value:.4f}")

    if p_value < significance_level:
        print("Reject the null hypothesis: There is a significant difference between the means of the two samples.")
    else:
        print("Fail to reject the null hypothesis: There is no significant difference between the means of the two samples.")
```

```python
if __name__ == "__main__":
    np.random.seed(0)
    sample1 = np.random.normal(loc=100, scale=15, size=50)
    sample2 = np.random.normal(loc=105, scale=15, size=50)
    independent_t_test(sample1, sample2)
```

7. Perform a paired sample T-test using Python and visualize the comparison results.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

def paired_sample_t_test(before, after, significance_level=0.05):
    t_statistic, p_value = stats.ttest_rel(before, after)

    print(f"T-statistic: {t_statistic:.4f}")
    print(f"P-value: {p_value:.4f}")

    if p_value < significance_level:
        print("Reject the null hypothesis: There is a significant difference between the before and after measurements.")
    else:
        print("Fail to reject the null hypothesis: There is no significant difference between the before and after measurements.")

    visualize_comparison(before, after)

def visualize_comparison(before, after):
    plt.figure(figsize=(8, 6))
    labels = ['Before', 'After']
    means = [np.mean(before), np.mean(after)]
    stds = [np.std(before), np.std(after)]

    plt.bar(labels, means, yerr=stds, capsize=10, color=['blue', 'orange'], alpha=0.7)
    plt.scatter(['Before'] * len(before), before, color='blue', alpha=0.6, label='Before measurements')
    plt.scatter(['After'] * len(after), after, color='orange', alpha=0.6, label='After measurements')
    plt.title('Comparison of Before and After Measurements (Paired Sample T-test)')
    plt.ylabel('Measurement Value')
    plt.legend()
    plt.show()
```

```python
if __name__ == "__main__":
    np.random.seed(0)
    before = np.random.normal(loc=80, scale=10, size=30)
    after = before + np.random.normal(loc=5, scale=5, size=30)

    paired_sample_t_test(before, after)
```

9. Write a Python function to calculate the confidence interval for a sample mean and explain its significance.

```python
import numpy as np
from scipy import stats

def confidence_interval(sample_data, confidence_level=0.95):
    sample_mean = np.mean(sample_data)
    sample_std = np.std(sample_data, ddof=1)
    sample_size = len(sample_data)
    z_score = stats.norm.ppf(1 - (1 - confidence_level) / 2)
    margin_of_error = z_score * (sample_std / np.sqrt(sample_size))

    lower_bound = sample_mean - margin_of_error
    upper_bound = sample_mean + margin_of_error

    return lower_bound, upper_bound

```

```python
if __name__ == "__main__":
    np.random.seed(0)
    sample_data = np.random.normal(loc=75, scale=10, size=50)
    lower_bound, upper_bound = confidence_interval(sample_data, confidence_level=0.95)
    print(f"95% Confidence Interval: ({lower_bound:.2f}, {upper_bound:.2f})")
```

10. Write a Python program to calculate the margin of error for a given confidence level using sample data.

```python
import math
from scipy.stats import norm

def calculate_margin_of_error(confidence_level, sample_std_dev, sample_size):
    z_score = norm.ppf(1 - (1 - confidence_level) / 2)
    standard_error = sample_std_dev / math.sqrt(sample_size)
    margin_of_error = z_score * standard_error

    return margin_of_error

if __name__ == "__main__":
    confidence_level = 0.95
    sample_std_dev = 12.5
    sample_size = 100

    margin_of_error = calculate_margin_of_error(confidence_level, sample_std_dev, sample_size)
    print(f"Margin of Error: {margin_of_error:.2f}")
```

11. Implement a Bayesian inference method using Bayes' Theorem in Python and explain the process.

```python
def bayesian_inference(prior, likelihood_given_h, likelihood_given_not_h):
    prior_not_h = 1 - prior
    numerator = likelihood_given_h * prior
    denominator = (likelihood_given_h * prior) + (likelihood_given_not_h * prior_not_h)
    posterior = numerator / denominator

    return posterior

if __name__ == "__main__":
    prior = 0.01
    likelihood_given_h = 0.95
    likelihood_given_not_h = 0.05

    posterior = bayesian_inference(prior, likelihood_given_h, likelihood_given_not_h)
    print(f"Posterior Probability (P(H|E)): {posterior:.4f}")
```

12. Perform a Chi-square test for independence between two categorical variables in Python.

```python
import numpy as np
import pandas as pd
from scipy.stats import chi2_contingency
data = np.array([[30, 20],
                 [20, 30]])

contingency_table = pd.DataFrame(data, columns=["Product A", "Product B"], index=["Male", "Female"])
print("Contingency Table:")
print(contingency_table)

chi2, p, dof, expected = chi2_contingency(data)

print("\nChi-square Test Results:")
print(f"Chi-square Statistic: {chi2:.4f}")
print(f"P-value: {p:.4f}")
print(f"Degrees of Freedom: {dof}")
print("\nExpected Frequencies:")
print(expected)

alpha = 0.05
if p < alpha:
    print("\nConclusion: There is a significant association between Gender and Preference.")
else:
    print("\nConclusion: There is no significant association between Gender and Preference.")
```

13. Write a Python program to calculate the expected frequencies for a Chi-square test based on observed
data.

```python
import numpy as np
import pandas as pd

def calculate_expected_frequencies(observed):
    observed = np.array(observed)
    row_totals = observed.sum(axis=1)
    col_totals = observed.sum(axis=0)
    total = observed.sum()
    expected = np.outer(row_totals, col_totals) / total
    return expected

observed_data = np.array([[50, 30, 20],  # Row 1
                          [30, 40, 30],  # Row 2
                          [20, 30, 50]]) # Row 3

observed_df = pd.DataFrame(observed_data, columns=["Category A", "Category B", "Category C"],
                           index=["Group 1", "Group 2", "Group 3"])
print("Observed Frequencies:")
print(observed_df)

expected_frequencies = calculate_expected_frequencies(observed_data)

expected_df = pd.DataFrame(expected_frequencies, columns=["Category A", "Category B", "Category C"],
                           index=["Group 1", "Group 2", "Group 3"])
print("\nExpected Frequencies:")
print(expected_df)
```

14. Perform a goodness-of-fit test using Python to compare the observed data to an expected distribution.

```python
import numpy as np
from scipy.stats import chisquare
observed = np.array([50, 30, 20])
expected_proportions = np.array([0.4, 0.4, 0.2])
total = observed.sum()
expected = expected_proportions * total
chi2_stat, p_value = chisquare(f_obs=observed, f_exp=expected)

print("Observed Frequencies:", observed)
print("Expected Frequencies:", expected)
print(f"\nChi-square Statistic: {chi2_stat:.4f}")
print(f"P-value: {p_value:.4f}")

alpha = 0.05
if p_value < alpha:
    print("\nConclusion: The observed data significantly differ from the expected distribution.")
else:
    print("\nConclusion: The observed data do not significantly differ from the expected distribution.")
```

16. Implement an F-test using Python to compare the variances of two random samples.

```python
import numpy as np
from scipy.stats import f

def f_test(sample1, sample2, alpha=0.05):
    var1 = np.var(sample1, ddof=1)
    var2 = np.var(sample2, ddof=1)
    f_stat = var1 / var2 if var1 > var2 else var2 / var1
    df1 = len(sample1) - 1
    df2 = len(sample2) - 1
    p_value = 2 * min(f.cdf(f_stat, df1, df2), 1 - f.cdf(f_stat, df1, df2))
    conclusion = "Fail to reject H0: Variances are equal"
    if p_value < alpha:
        conclusion = "Reject H0: Variances are significantly different"

    return {
        "F-statistic": f_stat,
        "p-value": p_value,
        "Conclusion": conclusion,
        "Degrees of Freedom": (df1, df2),
    }
```

```python
if __name__ == "__main__":
    np.random.seed(42)
    sample1 = np.random.normal(10, 5, 30)
    sample2 = np.random.normal(12, 8, 40)

    result = f_test(sample1, sample2)

    print("F-test Results:")
    print(f"F-statistic: {result['F-statistic']:.4f}")
    print(f"P-value: {result['p-value']:.4f}")
    print(f"Degrees of Freedom: {result['Degrees of Freedom']}")
    print(f"Conclusion: {result['Conclusion']}")
```

19. Write a Python function to check the assumptions (normality, independence, and equal variance) for ANOVA.

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import normaltest, levene, shapiro

def check_anova_assumptions(groups, group_labels):
    print("\nNormality Test Results:")
    for i, group in enumerate(groups):
        stat, p_value = normaltest(group)
        print(f"Group {group_labels[i]} (D'Agostino's test): p-value = {p_value:.4f}")

        stat, p_value_shapiro = shapiro(group)
        print(f"Group {group_labels[i]} (Shapiro-Wilk test): p-value = {p_value_shapiro:.4f}")

    print("\nEqual Variance Test (Levene's test):")
    stat, p_value_levene = levene(*groups)
    print(f"Levene's test p-value: {p_value_levene:.4f}")

    print("\nVisualizing the data distributions:")
    plt.figure(figsize=(12, 6))

    plt.subplot(1, 2, 1)
    sns.boxplot(data=groups)
    plt.xticks(range(1, len(group_labels) + 1), group_labels)
    plt.title("Boxplots for Equal Variance")

    plt.subplot(1, 2, 2)
    for i, group in enumerate(groups):
        sns.histplot(group, kde=True, label=group_labels[i], bins=15)
    plt.legend()
    plt.title("Histograms for Normality Check")
    plt.tight_layout()
    plt.show()

    alpha = 0.05
    print("\nAssumptions Check Conclusion:")

    normality_ok = all(p > alpha for group in groups for p in [normaltest(group)[1], shapiro(group)[1]])
    if normality_ok:
        print("Normality: All groups passed normality assumption.")
    else:
        print("Normality: One or more groups failed normality assumption.")

    if p_value_levene > alpha:
        print("Equal Variance: Groups passed the equal variance assumption.")
    else:
        print("Equal Variance: Groups failed the equal variance assumption.")

    print("Independence: Assumed to be satisfied based on study design (this should be checked in data collection).")
```

```python
if __name__ == "__main__":
    np.random.seed(42)
    group1 = np.random.normal(50, 5, 30)
    group2 = np.random.normal(55, 5, 30)
    group3 = np.random.normal(60, 5, 30)

    check_anova_assumptions([group1, group2, group3], group_labels=["Group 1", "Group 2", "Group 3"])
```

20. Perform a two-way ANOVA test using Python to study the interaction between two factors and visualize the
results.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from statsmodels.formula.api import ols
from statsmodels.stats.anova import anova_lm

def perform_two_way_anova(data, formula):
    model = ols(formula, data=data).fit()
    anova_table = anova_lm(model)
    print("\nTwo-Way ANOVA Results:")
    print(anova_table)

    if anova_table["PR(>F)"]["factor1"] < 0.05:
        print("Factor 1 is significant.")
    if anova_table["PR(>F)"]["factor2"] < 0.05:
        print("Factor 2 is significant.")
    if anova_table["PR(>F)"]["factor1:factor2"] < 0.05:
        print("The interaction between Factor 1 and Factor 2 is significant.")

    sns.set(style="whitegrid")
    plt.figure(figsize=(10, 6))
    sns.catplot(x='factor1', y='dependent', hue='factor2', kind='point', data=data, markers=["o", "s"], linestyles=["-", "--"])
    plt.title("Interaction Plot between Factor 1 and Factor 2")
    plt.xlabel("Factor 1")
    plt.ylabel("Dependent Variable")
    plt.show()
```

```python
if __name__ == "__main__":
    np.random.seed(42)
    factor1 = np.repeat(["A", "B"], 50)
    factor2 = np.tile(["X", "Y"], 50)
    dependent = np.random.normal(10, 5, 100) + np.where(factor1 == "A", 2, 0) + np.where(factor2 == "X", 3, 0)

    data = pd.DataFrame({
        'factor1': factor1,
        'factor2': factor2,
        'dependent': dependent
    })

    formula = 'dependent ~ factor1 * factor2'
    perform_two_way_anova(data, formula)
```

24.  Perform a hypothesis test for population variance using a Chi-square distribution and interpret the results.

```python
import numpy as np
from scipy.stats import chi2
import matplotlib.pyplot as plt

def perform_chi_square_variance_test(sample_data, population_variance, alpha=0.05):
    n = len(sample_data)
    sample_variance = np.var(sample_data, ddof=1)
    chi_square_statistic = (n - 1) * sample_variance / population_variance

    df = n - 1

    chi_square_lower = chi2.ppf(alpha / 2, df)
    chi_square_upper = chi2.ppf(1 - alpha / 2, df)

    print(f"Chi-Square Test Results:")
    print(f"Chi-Square Statistic: {chi_square_statistic:.4f}")
    print(f"Critical Values: Lower = {chi_square_lower:.4f}, Upper = {chi_square_upper:.4f}")

    if chi_square_statistic < chi_square_lower or chi_square_statistic > chi_square_upper:
        print(f"Since the Chi-square statistic is outside the critical values, we reject the null hypothesis.")
    else:
        print(f"Since the Chi-square statistic is within the critical values, we fail to reject the null hypothesis.")

    x = np.linspace(0, chi_square_upper + 5, 500)
    y = chi2.pdf(x, df)

    plt.figure(figsize=(8, 6))
    plt.plot(x, y, label=f'Chi-square Distribution (df={df})', color='blue')
    plt.fill_between(x, y, where=(x <= chi_square_lower) | (x >= chi_square_upper), color='red', alpha=0.5, label='Rejection Region')
    plt.axvline(chi_square_statistic, color='green', linestyle='dashed', linewidth=2, label=f'Statistic = {chi_square_statistic:.4f}')
    plt.axvline(chi_square_lower, color='black', linestyle='dashed', label=f'Critical Lower = {chi_square_lower:.4f}')
    plt.axvline(chi_square_upper, color='black', linestyle='dashed', label=f'Critical Upper = {chi_square_upper:.4f}')

    plt.title("Chi-Square Test for Population Variance")
    plt.xlabel("Chi-Square Statistic")
    plt.ylabel("Probability Density")
    plt.legend()
    plt.grid(True)
    plt.show()
```

```python
if __name__ == "__main__":
    np.random.seed(42)
    sample_data = np.random.normal(loc=50, scale=10, size=30)
    population_variance = 100
    perform_chi_square_variance_test(sample_data, population_variance, alpha=0.05)
```

25. Write a Python script to perform a Z-test for comparing proportions between two datasets or groups.

```python
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt

def perform_z_test_for_proportions(successes1, n1, successes2, n2, alpha=0.05):
    p1 = successes1 / n1
    p2 = successes2 / n2

    pooled_p = (successes1 + successes2) / (n1 + n2)
    z_statistic = (p1 - p2) / np.sqrt(pooled_p * (1 - pooled_p) * (1/n1 + 1/n2))

    p_value = 2 * (1 - stats.norm.cdf(abs(z_statistic)))

    print(f"Z-statistic: {z_statistic:.4f}")
    print(f"P-value: {p_value:.4f}")

    if p_value < alpha:
        print(f"Since the p-value is less than {alpha}, we reject the null hypothesis.")
    else:
        print(f"Since the p-value is greater than {alpha}, we fail to reject the null hypothesis.")

    z_values = np.linspace(-4, 4, 1000)
    pdf_values = stats.norm.pdf(z_values, loc=0, scale=1)

    plt.figure(figsize=(8, 6))
    plt.plot(z_values, pdf_values, label="Standard Normal Distribution")
    plt.fill_between(z_values, pdf_values, where=(z_values >= abs(z_statistic)), color='red', alpha=0.5, label="Rejection Region")
    plt.axvline(z_statistic, color='green', linestyle='dashed', label=f'Z-statistic = {z_statistic:.4f}')
    plt.title("Z-test for Comparing Proportions")
    plt.xlabel("Z-value")
    plt.ylabel("Probability Density")
    plt.legend()
    plt.grid(True)
    plt.show()
```

```python
if __name__ == "__main__":
    successes1 = 120
    n1 = 200
    successes2 = 100
    n2 = 180

    perform_z_test_for_proportions(successes1, n1, successes2, n2, alpha=0.05)
```

26.  Implement an F-test for comparing the variances of two datasets, then interpret and visualize the results.

```python
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt

def perform_f_test(data1, data2, alpha=0.05):
    n1 = len(data1)
    n2 = len(data2)
    var1 = np.var(data1, ddof=1)
    var2 = np.var(data2, ddof=1)
    f_statistic = var1 / var2 if var1 >= var2 else var2 / var1
    df1 = n1 - 1
    df2 = n2 - 1
    p_value = 2 * min(stats.f.cdf(f_statistic, df1, df2), 1 - stats.f.cdf(f_statistic, df1, df2))

    print(f"F-statistic: {f_statistic:.4f}")
    print(f"P-value: {p_value:.4f}")

    if p_value < alpha:
        print(f"Since the p-value is less than {alpha}, we reject the null hypothesis.")
    else:
        print(f"Since the p-value is greater than {alpha}, we fail to reject the null hypothesis.")

    x = np.linspace(0, 5, 1000)
    y = stats.f.pdf(x, df1, df2)

    plt.figure(figsize=(8, 6))
    plt.plot(x, y, label=f'F-distribution (df1={df1}, df2={df2})', color='blue')
    plt.fill_between(x, y, where=(x >= f_statistic), color='red', alpha=0.5, label='Rejection Region')
    plt.axvline(f_statistic, color='green', linestyle='dashed', linewidth=2, label=f'F-statistic = {f_statistic:.4f}')
    plt.title("F-test for Comparing Variances")
    plt.xlabel("F-value")
    plt.ylabel("Probability Density")
    plt.legend()
    plt.grid(True)
    plt.show()
```

```python
if __name__ == "__main__":
    np.random.seed(42)
    data1 = np.random.normal(loc=50, scale=10, size=100)
    data2 = np.random.normal(loc=50, scale=20, size=100)

    perform_f_test(data1, data2, alpha=0.05)
```
